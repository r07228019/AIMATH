```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 50, 50)
y = 1.2 * x + np.random.randn(50) * 8 # 正確答案，等一下跟下面的學習結果做比較
```

```python
plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression() # 學習機
```

```python
X = x.reshape(50, 1) # 深度學習常需做reshape，來達到這個形式！！！(三個變數就是一列有三行)
```

```python
reg.fit(X, y)
```

```python
print(reg.predict([[69]]))
Y = reg.predict(X) # Y是對於每個X的預測結果，原本的正確值是x, y
plt.scatter(x, y)
plt.plot(x, Y, 'r')
```

```python
x = np.linspace(0, 5, 200)
y = 1.2*x + 0.8 + 0.5*np.random.randn(200)
plt.scatter(x, y)
```

## 選出70%的資料當訓練資料，30%當驗證資料

```python
from sklearn.model_selection import train_test_split
```

```python
X = x.reshape(200, 1)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 9487)
plt.scatter(x_train, y_train)
```

```python
regr = LinearRegression()
regr.fit(x_train, y_train)
y_pred = regr.predict(x_test)
```

```python
plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred, 'r')
```
