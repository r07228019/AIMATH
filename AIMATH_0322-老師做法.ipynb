## 讀資料、處理欄位問題、選取要的部分

```python
import pandas as pd
import numpy as np
df = pd.read_csv('http://stats.moe.gov.tw/files/detail/107/107_student.csv')
df.columns = df.loc[1]
df = df.loc[2: ]
df = df.loc[(df['等級別'] == 'B 學士') & (df['日間∕進修別'] == 'D 日')] # 要括號！！！
df
```

## 把資料轉成純數字！

```python
import locale
# atof
locale.setlocale(locale.LC_NUMERIC, '')

df['男生計'] = df['男生計'].apply(locale.atof)
df['女生計'] = df['女生計'].apply(locale.atof)
df['女/男'] = df['女生計'] / df['男生計']
df
```

## 排序

```python
df2 = df.sort_values('女/男', ascending = False)
df2
```
