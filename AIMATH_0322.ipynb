```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('http://stats.moe.gov.tw/files/detail/107/107_student.csv')
df = df.loc[(df["等級別"] == "B 學士") & (df["日間∕進修別"] == "D 日")] # 記得要刮號，否則會出錯！！！
df = df.loc[df["日間∕進修別"] == "D 日"]
df["一年級男生"] = [int(e.replace(",", "")) for e in df["一年級男生"]]
df["一年級女生"] = [int(e.replace(",", "")) for e in df["一年級女生"]]
df
```

## 方法一

```python
df = df.set_index("學校名稱")
df["一年級男生"] / (df["一年級男生"] + df["一年級女生"]) * 100
```

## 方法二

```python
df_p = pd.pivot_table(data = df, values = ["一年級男生", "一年級女生"], index = "學校名稱", aggfunc = "sum")
df_p["一年級男生"] / (df_p["一年級男生"] + df_p["一年級女生"]) * 100 # 樞紐分析表過後會重新排序
```
