# 利用政府鄉鎮指標資料庫中的2012總統大選得票率及鄉鎮扶養比來進行迴歸分析
#### 藍營得票率：藍營得票數/有效票數
#### 扶養比：(14歲以下人口數 + 65歲以上人口數) / 15~64歲人口數

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
mpl.rc('font', family='Noto Sans CJK TC')
%matplotlib inline
```

### 資料含有「藍營得票率」及「扶養比」兩個欄位(已擷取過)

```python
df = pd.read_csv("C:/Users/USER/AIMATH/選舉扶養比.csv", encoding = 'big5', engine = 'python')
df.head(5)
```

```python
x = np.array(df['扶養比'])
y = np.array(df['藍營得票率'])
X = x.reshape(358, 1)
```

### 原始資料散佈圖

```python
plt.scatter(x, y)
plt.xlabel('扶養比')
plt.ylabel('藍營得票率')
plt.title('鄉鎮扶養比與藍營得票率之關係')
```

### 將60%的資料拿來訓練，40%的資料拿來驗證，並展示迴歸結果，將預測值與實際值進行比較

```python
regr = LinearRegression()
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state = 9487)
regr.fit(x_train, y_train)
y_pred = regr.predict(x_test)
```

```python
plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred, 'r')
plt.xlabel('扶養比')
plt.ylabel('藍營得票率')
plt.title('鄉鎮扶養比與藍營得票率之關係')
```

#### 總結：<br><br>透過上述的結果，可以發現雖然迴歸出來的結果呈現些微的負相關，扶養比愈高的鄉鎮似乎藍營得票率愈低，但整體而言，兩者的關聯性並不大，無法真的透過扶養比的高低來預測得票率。
