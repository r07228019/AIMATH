## 專案介紹：
若要對公共運輸系統有良好的規劃，了解其使用者的類型與旅行特徵就顯得極為重要。不同的使用者類型可能會對旅運服務有不同的使用需求，而若能將其掌握，將有助於營運單位日後針對路線、票價、服務等相關規劃，在滿足使用者的需求之餘，亦可將有限的資源最最有效的分配。因此，本次我們以通勤時間的乘客族群為例，利用深度學習的方式，讓電腦能夠自動判識甚麼樣的使用者交易紀錄屬於「通勤學生」、「上班族」及「其他族群」，並了解這類型使用者的乘坐特性。期待未來能將更此辨識方式延用至更廣、更不易判斷的類型，對於公共運輸使用者的區分有近一步的了解。

## 方法：
**（一）訓練資料篩選**<br> 
本次專案使用的資料為台北市政府所提供單日之悠遊卡交易資料，其中包含卡號（已加密）、進站時間、進站站點、出站時間、出站站點等資訊。我們先將捷運站編碼後，以 3/15（週四）的捷運進出站資料為例，根據以下條件來篩選通勤族群資料：
*   通勤學生族群只考慮國、高中生，因為其通勤模式較為固定。大學生每天上課時間不固定，故不納入考慮。
*   一般通勤模式都是平日早上從家裡出發，然後在 6:30 ~ 9:00 這段時間抵達學校或工作地點，下午放學或下班後的時間約為 16:00 ~ 21:00 。早上9點到下午3點半以及凌晨五點半以前的進站的記錄為通勤族的機率較低，因此我們先透過上述規則來區分出通勤時間的卡號來作為訓練資料。
*   針對通勤時間的卡號，我們會去除通勤紀錄只有一次或同站進出的資料。
*   假設通勤國高中生所使用的通勤工具只有捷運，捷運站到學校的交通方式只靠步行則學校和捷運站的距離必須在步行可到的範圍內。若該卡號抵達的捷運站半徑一公里內並無國高中，我們便將該卡號認定為非國高中學生族群。其中，捷運站在半徑一公里內沒有學校的站點有：竹圍、唭哩岸、大湖公園、頂埔及輔大站（圖二）。
*   一般通勤模式大多是從家裡出發到學校/工作地點，而且必定會從學校/工作地點返家，同個卡號理應要有兩次以上的通勤紀錄，且第一次進站（離家）和最後一次出站的站點（返家）、第一次出站（抵達學校/工作地點）和第二次進站（離開學校/工作地點）的站點理應相同。因此，我們篩選出第一次進站＝最後一次出站、第一次出站＝第二次進站的卡號，且上下午皆有旅次的卡號作為通勤族群的訓練資料。

**（二）測試資料選用**<br>
每間國高中幾乎都有發行學生悠遊卡，刷學生悠遊卡的族群會有很大機率為學生族群，因此我們選用合理旅次中卡種為 6 (學生卡) 的卡號資料作為通勤學生族群的測試資料，卡種為 1 (一般卡) 的卡號資料則作為上班族的測試資料。至於其他族群的測試資料，我們選用卡種為 2 (敬老1) , 3 (敬老2) , 4 (愛心) , 5 (愛心陪同) , 9 (優待卡) 的卡號。


**<center>圖二：捷運站一公里環域。其中灰黑色點為捷運站環域一公里內沒有學校的捷運站。</center>**

**（三）訓練神經網路**
*    經篩選後，通勤時間內共有 228890 筆訓練資料；其中，學生卡資料有 51610 筆、一般卡資料有 167156 筆、其他卡資料有 10124 筆。我們以這些篩選完的三組資料為通勤學生/上班族/其他族群的訓練資料，再去設計神經網路，並且嘗試不同的 activation, loss function 及 optimizer 函數以改善模型精確度。
*   在訓練資料選用上，我們分別從通勤時間內一般卡、學生卡和其他卡種的隨機挑選 7/10 作為訓練資料，剩下的 3/10 作為測試資料。而在函數選用上，由於我們訓練的重點在於分辨卡號是否為學生，變量不為連續而是分門別類，所以我們嘗試sigmoid,  relu, tanh, Leaky ReLU, Maxout 作為 activation function，並選擇準確度較高者。
*   由於三種類別資料彼此差異性大，可以分解為兩類兩類彼此比較，因此在loss function 我們選擇使用 binary_crossentropy 函數搭配 relu、sigmoid，並選擇準確度較高者。在優化器（ optimizer ）選用上，我們使用 Adam 來優化結果。並使用兩種深度、寬度不同的深度學習模型進型比較。

## 結果與討論

經過 20 次訓練後，該模型分辨三種不同族群的準確率高達 0.8262 ，代表通勤時間內這三種不同族群中在時空上的分佈可能存在明顯差異。以下我們將針對不同通勤族群於空間上和時間上的分佈差異進行探討：<br>

**（一）通勤時間中不同族群的空間分布差異**<br>
假設捷運站沿線的地區通勤學生比率一致，則捷運站附近學校人數較多者應會有較多的通勤學生。從捷運站附近國高中人數圖（圖三）的分佈來看，可以發現人數較多的捷運站為淡水、北投、士林、三和國中、南港展覽館、台北小巨蛋、大坪林、新莊、永安市場、大安等站。其中，作為通勤學生族群測試資料的學生卡運量較大的捷運站分布（圖四）則和圖三相近，熱區分佈於淡水、士林、松江南京、台北小巨站、善導寺、市政府、大安、公館等站。一般卡運量較大的捷運站（圖四）則分佈在台北車站、南京復興、市政府、世貿/101等站，和圖三中的捷運站熱區有明顯差異。這些捷運站附近較多為公司行號或是商業區。而其他卡種（圖五）與一般卡種的熱區分布相似，運量較大的捷運站為台北車站、南京復興、市政府、世貿/101等站。因此，上班族和其他族群的通勤特性差異應該在時間上較為顯著。

**（二）通勤時間中不同族群的時間分布差異**<br>
而透過將不同卡種出入站時間進行視覺化，也能發現其差異。如圖七是三種卡種第一次進站時間的分布圖，其中可以發現一般卡的數量佔了多數，學生卡約二到三成，剩下的少部分則是其他卡（如愛心卡、敬老卡、軍警等）。學生卡在實施記名政策後，雖可能仍有冒用、混用的情情，但情況應該較少，故我們假定學生卡即為學生，一般卡為普通上班族。透過圖七，可以發現一般卡的第一次進站時間多集中於八點左右，且具有明顯高峰，學生卡則比一般卡更早，集中於七點出頭，這也與真實情況中上課時間比上班時間早的現象相符。而在其他卡種的部分，其數量少得多，不過可發現雖然最高峰是在七點半左右，但無明顯的集中趨勢，數量分布上也較平均，顯示出這些卡種較無固定時間通勤需求的特徵。另一個比較值得注意的是其他卡在極早的時間（六點前）的分布比例是比前兩者都高的，這可能與老年使用者習慣早出有關。<br>
而在圖八最後一次進站時間的分布中，一般卡在晚上六點半時的數量達到高峰，學生卡則峰值出現的時間同樣較一般卡來得早，約五點半至六點間，這與我們一般認知的下班、下課時間亦頗吻合。學生卡另一個小高峰出現於九點到十點之間，這可能是課後補習所造成的結果，與台灣的補習文化有關。而其他卡種仍舊是在數量上較少、分布上較平均，無明顯的高低起伏變化。<br>
以上的結果，表示學生卡、一般卡、其他卡使用者之間的旅行型態在時間分布上也是具有不同特徵的，也解釋了為何model能夠透過交易的時間與地點，區分出學生卡族群。

## 結論

即便在通勤時間內的旅次中看似時間同質性高，但經過深度學習模型的訓練後，仍可有效分辨出學生族群、上班族和其他族群，準確率高達八成。從時空分佈可看出這三種族群仍有細微的差異。在空間分佈上，學生族群易出現在學生密度較高的地方；上班族和其他族群比較容易出現在商業區。而在時間分佈上，學生族群整體來說其往返的通勤時間都較一般上班族來的早，集中在早上七點與傍晚五點半附近，可能是和學校作息有關，同時也與一般認知相符；上班族則集中出現的時間則較晚，尖峰值大約出現在早上八點左右，回程則無明顯峰值；至於其他族群的時間分佈則較為分散，尖峰值較不明顯，且呈現較類似常態分佈。由本次的實驗結果可抵得知：機器學習可以準確分辨出不同類型使用者的乘坐特性，對於公共運輸使用者的區分，以及後續用以全面整握旅運情況、調整規劃方向等方面，都將是一大突破。<br>
不過，由於本專案選用的資料是經過篩選，所選出者皆為較符合我們一般對通勤者定義的卡號與旅次，故較無法真實辨別出「不符合一般通勤定義的通勤者」，例如每天固定搭乘一次捷運、或主要通勤時間為夜間者。另外，也因為我們的資料只有捷運的部分，但一個城市內的公共運輸其實需要不同運具間的配合，才能滿足整體使用者的需求，因此在無法有效與其他運具如公車、YouBike做連結的情況下，對於族群的識別及解釋上可能會缺陷，故若將來能蒐集到其他運具的交易紀錄來整合分析，應該在實用性及正確率上都能獲得更好的結果。
